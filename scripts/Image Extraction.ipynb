```python
# including crop area while loop TWO pictures per x straid 

import os
import sys
from PIL import Image

pathIn = "/Users/nijeri/Documents/HSLU/4 Semester/Master Thesis/Pictures/Video/Video 1/Video 1\\frame3 .jpg"
pathOut = "/Users/nijeri/Documents/HSLU/4 Semester/Master Thesis/Pictures/Pictures/Video 1/"

#start and end points of landing pad --> is static for each video since camera was not moved
start_x = 970
end_x = 1670
y_1 = 415
y_2 = 475
y_3 = 475
y_4 = 535

# Crops the image and saves it as "new_filename"
def crop_image(img, crop_area, new_filename):
    cropped_image = img.crop(crop_area)
    cropped_image.save(new_filename)
    
    
# Creates the crop areas based on the landing pad pixels
def crop_area(start_x, end_x, y_1, y_2, y_3, y_4):
    crop_areas = []
    while start_x < end_x:
        crop_areas.append((start_x, y_1 , start_x + 60, y_2))
        crop_areas.append((start_x, y_3 , start_x + 60, y_4))
        start_x += 60
    return crop_areas

# The x, y coordinates of the areas to be cropped. (x1, y1, x2, y2)
crop_areas = crop_area(start_x, end_x, y_1, y_2, y_3, y_4)

image_name = pathIn
img = Image.open(image_name)

# Loops through the "crop_areas" list and crops the image based on the coordinates in the list
for i, crop_area in enumerate(crop_areas):
    filename = pathOut
    new_filename = filename + '_cropped' + str(i) +'.jpg'
    #print(new_filename)
    crop_image(img, crop_area, new_filename)
```

```python
# including crop area while loop ONE picture per x straid with for loop which loops trough all frames in folder

import os
import sys
from PIL import Image

videoNr = "Video 16"
pathOut = "/Users/nijeri/Documents/HSLU/4 Semester/Master Thesis/Pictures/Pictures/"+videoNr+"/"

#start and end points of landing pad --> is static for each video since camera was not moved
start_x = 740
end_x = 1370
y_1 = 665
y_2 = 725


# Crops the image and saves it as "new_filename"
def crop_image(img, crop_area, new_filename):
    cropped_image = img.crop(crop_area)
    cropped_image.save(new_filename)
    
    
# Creates the crop areas based on the landing pad pixels
def crop_area(start_x, end_x, y_1, y_2):
    crop_areas = []
    while start_x < end_x:
        crop_areas.append((start_x, y_1 , start_x + 60, y_2))
        start_x += 60
    return crop_areas

# The x, y coordinates of the areas to be cropped. (x1, y1, x2, y2)
crop_areas = crop_area(start_x, end_x, y_1, y_2)

# Defines which frames should be included
first_frame = 0
last_frame = 356

while first_frame <= last_frame:
    image_name = "/Users/nijeri/Documents/HSLU/4 Semester/Master Thesis/Pictures/Video/"+videoNr+"/"+videoNr+"\\frame"+str(first_frame)+".jpg"
    img = Image.open(image_name)

    # Loops through the "crop_areas" list and crops the image based on the coordinates in the list
    for i, crop_area in enumerate(crop_areas):
        filename = pathOut
        new_filename = filename + image_name.split("\\")[1].split(".")[0] + '_cropped' + str(i) +'.jpg'
        crop_image(img, crop_area, new_filename)
    first_frame += 1
```
